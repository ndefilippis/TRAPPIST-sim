# Secular Multiples and TRAPPIST-sim Tutorial

To run SecularMultiples on a particular bound system, we can use the following code snippet

```python
from amuse.lab import *
from amuse.plot import *
from sim_secularmultiples import run_secular_multiples


# Set up particle set, from the AMUSE textbook
bodies = Particles(3)

sun = bodies[0]
sun.mass = 1.0 | units.MSun
sun.radius = 1.0 | units.RSun
sun.position = (855251, -804836, 3186) | units.km
sun.velocity = (7.893, 11.894, 0.20642) | (units.m/units.s)

venus = bodies[1]
venus.mass = 0.0025642 | units.MJupiter
venus.radius = 3026.0 | units.km
venus.position = (-0.3767, 0.60159, 0.03930) | units.AU
venus.velocity = (-29.7725, -18.849, 0.795) | units.kms

earth = bodies[2]
earth.mass = 1.0 | units.MEarth
earth.radius = 1.0 | units.REarth
earth.position = (-0.98561, 0.0762, -7.847e-5) | units.AU
earth.velocity = (-2.927, -29.803, -0.0005327) | units.kms

bodies.move_to_center()

# Run secular multiples for 1 megayear, taking 100 samples
time, semimajor_axis, eccentricity, inclination, bodies = run_secular_multiples(bodies, 1.0 | units.Myr, 200)

# Plot orbital parameters over time
plot(time, eccentricity[0], time, eccentricity[1])
xlabel("Time")
ylabel("Eccentricity")
```
